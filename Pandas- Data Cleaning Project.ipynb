#Dataset information 

- tradeTime: the time of transaction
- DOM: active days on market
- followers: the number of people follow the transaction.
- totalPrice: the total price (measured in millions of RMB)
- price: the average price by square 
- square: the square of house
- bedroom: the number of bedroom
- drawingRoom: the number of drawing room
- kitchen: the number of kitchen
- bathroom the number of bathroom
- buildingType: including tower( 1 ) , bungalow( 2 )ï¼Œcombination of plate and tower( 3 ), plate( 4 ).
- constructionTime: the time of construction
- renovationCondition: including other( 1 ), rough( 2 ),Simplicity( 3 ), hardcover( 4 )
- buildingStructure: including unknow( 1 ), mixed( 2 ), brick and wood( 3 ), brick and concrete( 4 ),steel( 5 ) and steel-concrete composite ( 6 ). 
- elevator: elevator have ( 1 ) or not have elevator( 0 )
- fiveYearsProperty: if the owner have the property for less than 5 years, indicating whether the property is newly build or owned previously

```python
""""In the decriptive statistics above,buildingType has a minimum value of 0.048. This min value doesn't make any sense since 
buildingType is a categorical data ranges from 1 to 4. In other words, the min value of buildingType should be 1."""

#Filtering data and removing rows where buildingType < 1
df = df.drop(df[df.buildingType < 1].index)
df.drop(df[df.buildingType < 1].index, inplace = True)

#Checking the min value of buidlingType
df.describe()   #Or use df.loc[df['buildingType'] < 1] 
                #Now the min is 1   
```

```python
#Converting object to date in column tradeTime

df['tradeTime'] = df['tradeTime'].astype('datetime64[ns]')
#print(df['tradeTime'].dtypes)


#Converting object to int in columns bedroom, drawingRoom,and bathRoom

df[['bedroom','drawingRoom', 'bathRoom']] = df[['bedroom','drawingRoom', 'bathRoom']].apply(pd.to_numeric)
cols = ['elevator', 'fiveYearsProperty','subway']
df[cols] = df[cols].applymap(np.int64)
#df.info()


#Converting columns DOM, buildingType, and constructionTime with data type float NaN to integer

#print(df[df['DOM'].isnull()])  #Identify NaN values using boolean indexing
df['DOM'] = pd.to_numeric(df['DOM'], errors='coerce') #Removing all non-numeric values, and replace non-numeric values to NaNs
df = df.dropna(subset=['DOM']) #Remove all rows with NaNs in column DOM
df['DOM'] = df['DOM'].astype('int64')  #convert values to int

#print(df[df['buildingType'].isnull()])  #Identify NaN values using boolean indexing
df['buildingType'] = pd.to_numeric(df['buildingType'], errors='coerce') #Removing all non-numeric values, and replace non-numeric values to NaNs
df = df.dropna(subset=['buildingType']) #Remove all rows with NaNs in column buildingType
df['buildingType'] = df['buildingType'].astype('int64')  #convert values to int

#print(df[df['constructionTime'].isnull()])  #Identify NaN values using boolean indexing
df['constructionTime'] = pd.to_numeric(df['constructionTime'], errors='coerce') #Removing all non-numeric values, and replace non-numeric values to NaNs
df = df.dropna(subset=['constructionTime']) #Remove all rows with NaNs in column constructionTime
df['constructionTime'] = df['constructionTime'].astype('int64')  #convert values to int


#Checking data type
df.info()


#Now we have 152396 rows(and none of them is null!)
```
